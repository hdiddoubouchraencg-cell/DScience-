```python
# Install the ucimlrepo library to fetch datasets from UCI Machine Learning Repository
!pip install ucimlrepo
from ucimlrepo import fetch_ucirepo

# Fetch the Wine Quality dataset using its UCI ID (186)
wine_quality = fetch_ucirepo(id=186)

# Extract features (X) and target (y) into pandas DataFrames
X = wine_quality.data.features
y = wine_quality.data.targets

# Print metadata and variable information for the dataset
print(wine_quality.metadata)
print(wine_quality.variables)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Combine the features (X) and the target (y) into a single pandas DataFrame.
# This makes it easier to work with the data for plotting and analysis.
df = pd.concat([X, y], axis=1)
```

### Distribution of Wine Quality

```python
# Set the figure size for better visualization
plt.figure(figsize=(8, 6))

# Create a countplot to show the distribution of wine quality scores.
# 'x' is the quality score, 'data' is our DataFrame 'df'.
# 'palette' sets the color scheme, 'hue' is set to 'quality' for distinct colors per bar, and 'legend' is turned off.
sns.countplot(x='quality', data=df, palette='viridis', hue='quality', legend=False)

# Add a title to the plot
plt.title('Distribution of Wine Quality Scores')
# Label the x-axis
plt.xlabel('Quality Score')
# Label the y-axis
plt.ylabel('Number of Wines')

# Display the plot
plt.show()
```

### Distributions of Selected Physicochemical Properties

```python
# Create a figure with a 2x2 grid of subplots for multiple histograms
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Plot histogram for 'fixed_acidity'
sns.histplot(df['fixed_acidity'], kde=True, ax=axes[0, 0], color='skyblue')
axes[0, 0].set_title('Distribution of Fixed Acidity')
axes[0, 0].set_xlabel('Fixed Acidity')
axes[0, 0].set_ylabel('Count')

# Plot histogram for 'volatile_acidity'
sns.histplot(df['volatile_acidity'], kde=True, ax=axes[0, 1], color='lightcoral')
axes[0, 1].set_title('Distribution of Volatile Acidity')
axes[0, 1].set_xlabel('Volatile Acidity')
axes[0, 1].set_ylabel('Count')

# Plot histogram for 'citric_acid'
sns.histplot(df['citric_acid'], kde=True, ax=axes[1, 0], color='lightgreen')
axes[1, 0].set_title('Distribution of Citric Acid')
axes[1, 0].set_xlabel('Citric Acid')
axes[1, 0].set_ylabel('Count')

# Plot histogram for 'alcohol'
sns.histplot(df['alcohol'], kde=True, ax=axes[1, 1], color='gold')
axes[1, 1].set_title('Distribution of Alcohol Content')
axes[1, 1].set_xlabel('Alcohol Content')
axes[1, 1].set_ylabel('Count')

# Adjust layout to prevent titles and labels from overlapping
plt.tight_layout()
# Display the plots
plt.show()
```

### Relationship between Physicochemical Properties and Wine Quality

```python
# Create a figure with a 2x2 grid of subplots for multiple box plots
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Box plot for Fixed Acidity vs. Quality
sns.boxplot(x='quality', y='fixed_acidity', data=df, ax=axes[0, 0], palette='viridis', hue='quality', legend=False)
axes[0, 0].set_title('Fixed Acidity vs. Quality')
axes[0, 0].set_xlabel('Quality Score')
axes[0, 0].set_ylabel('Fixed Acidity')

# Box plot for Volatile Acidity vs. Quality
sns.boxplot(x='quality', y='volatile_acidity', data=df, ax=axes[0, 1], palette='plasma', hue='quality', legend=False)
axes[0, 1].set_title('Volatile Acidity vs. Quality')
axes[0, 1].set_xlabel('Quality Score')
axes[0, 1].set_ylabel('Volatile Acidity')

# Box plot for Citric Acid vs. Quality
sns.boxplot(x='quality', y='citric_acid', data=df, ax=axes[1, 0], palette='magma', hue='quality', legend=False)
axes[1, 0].set_title('Citric Acid vs. Quality')
axes[1, 0].set_xlabel('Quality Score')
axes[1, 0].set_ylabel('Citric Acid')

# Box plot for Alcohol vs. Quality
sns.boxplot(x='quality', y='alcohol', data=df, ax=axes[1, 1], palette='cividis', hue='quality', legend=False)
axes[1, 1].set_title('Alcohol vs. Quality')
axes[1, 1].set_xlabel('Quality Score')
axes[1, 1].set_ylabel('Alcohol Content')

# Adjust layout to prevent titles and labels from overlapping
plt.tight_layout()
# Display the plots
plt.show()
```
